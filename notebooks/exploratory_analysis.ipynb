# Sales Performance Exploratory Analysis

This notebook explores the cleaned sales data and KPI exports generated by `python src/main.py`.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
processed = PROJECT_ROOT / 'data' / 'processed'

sales = pd.read_csv(processed / 'cleaned_sales.csv')
monthly = pd.read_csv(processed / 'monthly_revenue_summary.csv')
products = pd.read_csv(processed / 'product_performance.csv')
categories = pd.read_csv(processed / 'category_performance.csv')
regions = pd.read_csv(processed / 'regional_performance.csv')
channels = pd.read_csv(processed / 'sales_channel_performance.csv')

sales.head()
```

## Dataset Health

```python
summary = {
    'rows': len(sales),
    'columns': sales.shape[1],
    'total_revenue': round(sales['net_revenue'].sum(), 2),
    'total_orders': sales['transaction_id'].nunique(),
    'date_min': sales['transaction_date'].min(),
    'date_max': sales['transaction_date'].max(),
}
summary
```

## Monthly Revenue Trend

```python
ax = monthly.plot(x='transaction_month', y='total_revenue', figsize=(14, 5), legend=False, marker='o')
ax.set_title('Monthly Revenue Trend')
ax.set_xlabel('Month')
ax.set_ylabel('Net Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
```

## Product, Category, Region, And Channel Leaders

```python
leaders = pd.DataFrame({
    'segment': ['Product', 'Category', 'Region', 'Sales Channel'],
    'leader': [products.iloc[0]['product_name'], categories.iloc[0]['category'], regions.iloc[0]['region'], channels.iloc[0]['sales_channel']],
    'revenue': [products.iloc[0]['total_revenue'], categories.iloc[0]['total_revenue'], regions.iloc[0]['total_revenue'], channels.iloc[0]['total_revenue']],
})
leaders
```

```python
products.head(10).plot(kind='barh', x='product_name', y='total_revenue', figsize=(10, 6), legend=False)
plt.title('Top 10 Products By Revenue')
plt.xlabel('Net Revenue')
plt.ylabel('Product')
plt.gca().invert_yaxis()
plt.tight_layout()
```
